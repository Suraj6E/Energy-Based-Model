```python
import numpy as np
import plotly.graph_objs as go
# from plotly.offline import plot



# Extract coordinates
x_coords = np.array([1,2,4])
y_coords = np.array([2,1,1])
z_coords = np.array([3,3,3])

# Create a grid
x_grid, y_grid = np.meshgrid(np.linspace(min(x_coords), max(x_coords), 100),
                              np.linspace(min(y_coords), max(y_coords), 100))
z_grid = np.zeros_like(x_grid)

# Interpolate z-values for each point on the grid
for i in range(x_grid.shape[0]):
    for j in range(x_grid.shape[1]):
        distances = np.sqrt((x_coords - x_grid[i, j])**2 + (y_coords - y_grid[i, j])**2)
        closest_point_idx = np.argmin(distances)
        z_grid[i, j] = z_coords[closest_point_idx]

# Create a Surface trace
trace = go.Surface(
    x=x_grid,
    y=y_grid,
    z=z_grid,
)

# Create layout
layout = go.Layout(
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z')
    ),
    title='Mesh Points Visualization (Surface)',
    width=1200,  # Set the width of the plot
    height=800,  # Set the height of the plot
    margin=dict(l=0, r=0, b=0, t=0)  # Set the margin of the plot
)

# Create figure
fig = go.Figure(data=[trace], layout=layout)

fig.show()
```
